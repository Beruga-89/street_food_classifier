# Define constant, functions and clear old data

Define a utility function to set random seeds for reproducibility across Python, NumPy, and PyTorch (CPU and GPU).
Also configure CUDA to operate deterministically.

Set global variables for directories to save models and sample images, dataset folder path,
and filenames for saving training history and best model checkpoints.

```python
def seed_everything(seed):
    # Set fixed seed for reproducibility across libraries
    random.seed(seed)                          # Python random module seed
    np.random.seed(seed)                       # NumPy seed
    torch.manual_seed(seed)                    # PyTorch CPU seed
    torch.cuda.manual_seed(seed)               # PyTorch CUDA seed for single GPU
    torch.cuda.manual_seed_all(seed)           # PyTorch CUDA seed for all GPUs if using multi-GPU
    # Ensure deterministic behavior for CUDA convolution operations
    torch.backends.cudnn.deterministic = True
    # Disable benchmark mode to prevent nondeterministic algorithm selection
    torch.backends.cudnn.benchmark = False
    os.environ["PYTHONHASHSEED"] = str(seed)
    logging.basicConfig(filename="training.log", level=logging.INFO)
    logging.info("Seed gesetzt")


# Directory to save trained models
MODEL_FOLDER = "models"

# Directory to save sample images for visualization or debugging
SAMPLE_FOLDER = "sample"

# Flag to indicate whether to clear old data before running
clear = True

# Path to dataset folder containing images organized by class
DATA_FOLDER = Path('popular_street_foods/dataset')

# File path to save plots (e.g., training/validation loss or accuracy graphs)
plot_image_path: str = 'plot_image.png'

# File path to save training history (loss, accuracy per epoch) in JSON format
history_path: str = "history.json"

# File path to save the best model checkpoint according to F1-score
model_path: str = 'best_f1_model.pth'
```

## visualize dataset (train and val) label

Define a utility function to extract all labels from a DataLoader to analyze label distribution.
Use it to count labels in training and validation sets.
Visualize label distribution stacked to compare train and validation class balance.

```python
def count(data_loader: DataLoader):
    real_labels = []
    for images, labels in tqdm(data_loader, desc="Counting", leave=False):
        real_labels.extend(labels.cpu().numpy().tolist())
    return np.array(real_labels)


def eda(train_data: np.ndarray, val_data: np.ndarray, title='Stacked Label Distribution'):
    train_counts = Counter(train_data)
    val_counts = Counter(val_data)

    all_labels = sorted(set(train_counts.keys()) | set(val_counts.keys()))

    train_values = [train_counts.get(l, 0) for l in all_labels]
    val_values = [val_counts.get(l, 0) for l in all_labels]

    x = np.arange(len(all_labels))

    plt.figure(figsize=(8, 5))
    plt.bar(x, train_values, label='Train', color='skyblue')
    plt.bar(x, val_values, bottom=train_values, label='Validation', color='orange')

    plt.xticks(x, all_labels)
    plt.xlabel('Label')
    plt.ylabel('Count')
    plt.title(title)
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

val_labels = count(val_loader)
train_labels = count(train_loader)

eda(train_labels, val_labels, title='Train/Validation Label Distribution')
```

# Define helper functions

## Evaluate function
Define evaluation function to run model in evaluation mode on validation data.
Compute average loss, accuracy, and weighted F1-score.

```python
@torch.no_grad()
def evaluate(model: nn.Module, data_loader: DataLoader, 
             device: torch.device, criterion: F):
    model.eval()
    all_preds = []
    all_labels = []
    total_loss = 0

    for images, labels in tqdm(data_loader, desc="Evaluating", leave=False):
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        loss = criterion(outputs, labels)
        total_loss += loss.item() * images.size(0)
        preds = outputs.argmax(dim=1).cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(labels.cpu().numpy())

    avg_loss = total_loss / len(data_loader.dataset)
    accuracy = accuracy_score(all_labels, all_preds)
    f1 = f1_score(all_labels, all_preds, average='weighted')
    return {'loss': avg_loss, 'accuracy': accuracy, 'f1': f1}
```

## Visualize history function
This function visualizes training and validation metrics (like loss or accuracy) over epochs. It creates subplots for each metric, optionally saves the plot and training history as files, and then displays the plot.

```python
def plot_history(history: dict[str, dict[str, list[float]]],
                 plot_image_path: str = 'plot_image.png',
                 history_path: str = "history.json",
                 save: bool = True):
    # Extract training and validation history
    train_history = history['train']
    val_history = history['val']
    
    # Generate epoch indices
    epochs = range(1, len(train_history['loss']) + 1)
    
    # Define full paths for saving plot image and history file
    history_path = os.path.join(SAMPLE_FOLDER, history_path)
    plot_image_path = os.path.join(SAMPLE_FOLDER, plot_image_path)
    
    # Create a figure with subplots for each metric
    plt.figure(figsize=(14, 5))
    
    # Loop through each metric (e.g., loss, accuracy)
    for i, k in enumerate(train_history):
        plt.subplot(1, 3, i + 1)  # Create a subplot
        plt.plot(epochs, train_history[k], 'bo-', label=f'Train {k.capitalize()}')  # Training curve
        plt.plot(epochs, val_history[k], 'ro-', label=f'Val {k.capitalize()}')      # Validation curve
        plt.xlabel('Epoch')
        plt.ylabel(k.capitalize())
        plt.title(f'{k.capitalize()} over Epochs')
        plt.legend()
        plt.grid(True)

    plt.tight_layout()
    
    # Save plot and history if requested
    if save:
        plt.savefig(plot_image_path)
        with open(history_path, 'w') as f:
            json.dump(history, f, indent=4)
    
    # Show the plot
    plt.show()
```

## Get lastest checkpoint model
This function scans a folder to find model checkpoint files named with a specific pattern (e.g., `model_epoch_10.pth`) and returns the highest epoch number found. Useful for resuming training from the latest checkpoint.

```python
def get_latest_epoch(folder: str = './', prefix="model_epoch_", suffix=".pth"):
    max_epoch = 0  # Initialize max epoch to 0

    # Iterate through all files in the specified folder
    for filename in os.listdir(folder):
        # Check if the filename matches the pattern "model_epoch_{number}.pth"
        match = re.match(fr"{prefix}(\d+){suffix}", filename)
        if match:
            # Extract the epoch number from the filename
            epoch_num = int(match.group(1))
            # Update max_epoch if this epoch number is greater
            if epoch_num > max_epoch:
                max_epoch = epoch_num

    return max_epoch  # Return the highest epoch number found
```

## Define training functions

Define training loop function for one epoch.
Run model training mode, calculate loss and update weights.
Track loss, accuracy, and weighted F1-score for each batch.

```python
def train(model: nn.Module, train_loader: DataLoader, 
          optimizer: torch.optim.Adam, device: torch.device,
          criterion: F):
    model.train()  # Set model to training mode

    losses = []        # To store loss for each batch
    all_preds = []     # To accumulate predictions
    all_labels = []    # To accumulate true labels

    correct_train = 0  # Correct predictions counter
    total_train = 0    # Total samples counter

    # Progress bar for training
    loop = tqdm(train_loader, desc="Training", leave=False)
    for images, labels in loop:
        images, labels = images.to(device), labels.to(device)

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)
        losses.append(loss.item())

        # Get predicted class labels
        _, predicted = torch.max(outputs, 1)

        # Update accuracy counters
        correct_train += (predicted == labels).sum().item()
        total_train += labels.size(0)

        # Backpropagation and optimizer step
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        # Store predictions and labels for F1 score
        all_preds.extend(predicted.cpu())
        all_labels.extend(labels.cpu())

        # Compute batch metrics for progress display
        accuracy = (predicted == labels).sum().item() / labels.size(0)
        f1 = f1_score(labels.cpu(), predicted.cpu(), average='weighted')

        # Update progress bar with current metrics
        loop.set_postfix(loss=loss.item(), accuracy=accuracy, f1=f1)

    # Compute epoch-level average metrics
    avg_loss = np.mean(losses)
    accuracy = correct_train / total_train
    f1 = f1_score(all_labels, all_preds, average='weighted')

    return {'loss': avg_loss, 'accuracy': accuracy, 'f1': f1}
# Best Practice: EarlyStopping und LearningRateScheduler sollten hier ergänzt werden
```

## Define check point saving functions

* `save_best_models()` checks if the current model has the best validation loss, accuracy, or F1 score so far. If so, it saves the model and updates the corresponding best metric.

* `save_epoch_model()` saves the model at the current epoch, regardless of performance, for record keeping or potential rollback.

```python
def save_best_models(model: nn.Module, 
                     val_result: dict, epoch: int, 
                     loss_save_path, acc_save_path, f1_save_path):
    global best_loss, best_accuracy, best_f1
    updated = False  # Track whether any model was updated

    # Build full paths for saving model checkpoints
    loss_save_path = os.path.join(MODEL_FOLDER, loss_save_path)
    acc_save_path = os.path.join(MODEL_FOLDER, acc_save_path)
    f1_save_path = os.path.join(MODEL_FOLDER, f1_save_path)

    # Check if current validation loss is the best so far
    if val_result['loss'] < best_loss:
        best_loss = val_result['loss']
        torch.save(model.state_dict(), loss_save_path)
        print(f"✅Best loss model saved at epoch {epoch + 1:02d}")
        updated = True

    # Check if current validation accuracy is the best so far
    if val_result['accuracy'] > best_accuracy:
        best_accuracy = val_result['accuracy']
        torch.save(model.state_dict(), acc_save_path)
        print(f"✅Best accuracy model saved at epoch {epoch + 1:02d}")
        updated = True

    # Check if current validation F1 score is the best so far
    if val_result['f1'] > best_f1:
        best_f1 = val_result['f1']
        torch.save(model.state_dict(), f1_save_path)
        print(f"✅Best F1 model saved at epoch {epoch + 1:02d}")
        updated = True

    return updated  # Return whether any model was updated


def save_epoch_model(model: nn.Module, epoch: int):
    # Save the model for the current epoch (regardless of performance)
    path = os.path.join(MODEL_FOLDER, f"model_epoch_{epoch + 1:02d}.pth")
    torch.save(model.state_dict(), path)
    print(f"Model for epoch {epoch + 1:02d} saved.")
```

## Fit function
This function trains a model over multiple epochs with early stopping and learning rate scheduling. It saves the best models based on validation loss, accuracy, and F1 score, logs training progress, and returns the full training history. It also supports resuming from the last saved epoch.

```python
def fit(model: nn.Module, optimizer: torch.optim.Adam,
        device: torch.device, epochs: int, 
        train_loader: DataLoader, val_loader: DataLoader, criterion: F,
        gamma: float = 0.1, patience = 5,
        loss_save_path = 'best_loss_model.pth', 
        acc_save_path = 'best_acc_model.pth', 
        f1_save_path = 'best_f1_model.pth',
        history: dict[str, dict[str, list[float]]] = {}):
    
    global best_loss, best_accuracy, best_f1

    # Resume training from the last saved epoch
    start_epoch = get_latest_epoch(MODEL_FOLDER)

    # Learning rate scheduler: reduce LR if F1 score plateaus
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='max', factor=gamma, patience=3)

    early_stop_counter = 0          # Early stopping counter
    lr_no_improve_counter = 0       # Learning rate plateau counter

    for epoch in range(start_epoch, start_epoch + epochs):
        start_time = time.time()
        print('=' * 25 + f'Epoch {epoch + 1:02d}/{start_epoch + epochs:02d}' + '=' * 25)

        # Train for one epoch
        train_result = train(model, train_loader, optimizer, device, criterion)
        # Evaluate on validation data
        val_result = evaluate(model, val_loader, device, criterion)

        # Update training history
        history['train']['loss'].append(train_result['loss'])
        history['train']['accuracy'].append(train_result['accuracy'])
        history['train']['f1'].append(train_result['f1'])

        history['val']['loss'].append(val_result['loss'])
        history['val']['accuracy'].append(val_result['accuracy'])
        history['val']['f1'].append(val_result['f1'])

        # Save model if it's the best so far
        if save_best_models(model, val_result, epoch, loss_save_path, acc_save_path, f1_save_path):
            early_stop_counter = 0
            lr_no_improve_counter = 0
        else:
            early_stop_counter += 1
            lr_no_improve_counter += 1
            print(f'🚫🚫Stopping counter {early_stop_counter}/{patience}')
            print(f'🚫🚫Learning rate counter {lr_no_improve_counter}/{3}')

        # Always save model for the current epoch
        save_epoch_model(model, epoch)

        # Log epoch duration
        minutes, seconds = divmod(time.time() - start_time, 60)
        print(f"Epoch time: {int(minutes):02d}:{int(seconds):02d} (mm:ss)")

        # Print performance metrics
        current_lr = optimizer.param_groups[0]['lr']
        print(f"Epoch [{epoch+1}/{start_epoch + epochs}]"
              f"Train Loss: {train_result['loss']:.4f} | "
              f"Val Loss: {val_result['loss']:.4f} | "
              f"Train Acc: {train_result['accuracy']:.4f} | "
              f"Val Acc: {val_result['accuracy']:.4f} | "
              f"Train F1: {train_result['f1']:.4f} | "
              f"Val F1: {val_result['f1']:.4f} | "
              f"LR: {current_lr:.6f}")

        # Early stopping condition
        if early_stop_counter >= patience:
            print("❌❌Early stopping triggered")
            break

        # Step scheduler based on validation F1
        scheduler.step(val_result['f1'])

        # Log learning rate change if triggered
        if lr_no_improve_counter >= 3:
            current_lr = optimizer.param_groups[0]['lr']
            print(f"🆎🆎Learning rate reduced to {current_lr:.6f} at epoch {epoch + 1:02d}")
            lr_no_improve_counter = 0

    return history  # Return complete training history
# Best Practice: EarlyStopping und LearningRateScheduler sollten hier ergänzt werden
```

## Load best state model
This snippet loads a pre-trained model’s weights from disk. It automatically handles whether to load them onto the GPU or CPU based on availability and ensures the model is transferred to the correct device for further use.

```python
# Build full path to the model file
model_path = os.path.join(MODEL_FOLDER, model_path)

# Check if the model file exists
if os.path.exists(model_path):
    if torch.cuda.is_available():
        # Load model weights directly to GPU if available
        model.load_state_dict(torch.load(model_path, weights_only=True))
    else:
        # Load model weights and map to CPU if GPU is unavailable
        model.load_state_dict(torch.load(model_path, weights_only=True, map_location=torch.device('cpu')))

    # Move model to the correct device (GPU or CPU)
    model.to(device)
```

## Define predict function
This function takes either a file path or a NumPy array as input, processes the image, and returns the predicted class label if the confidence is above the threshold. Otherwise, it returns a default label like `"unknown"`. It handles grayscale and single-channel images and supports both CPU and GPU execution.

```python
def predict_image(source: str | np.ndarray, threshold=0.4, unknown_label="unknown"):
    model.eval()  # Set model to evaluation mode

    # Load image from file path
    if isinstance(source, str):
        image_path = source.strip()
        if not os.path.exists(image_path):
            raise FileNotFoundError(f"{image_path} does not exist")
        image = Image.open(image_path).convert('RGB')

    # Handle image from numpy array input
    elif isinstance(source, np.ndarray):
        if source.ndim == 2:
            # Grayscale → RGB
            source = np.stack([source] * 3, axis=-1)
        elif source.ndim == 3 and source.shape[2] == 1:
            # Single-channel → RGB
            source = np.repeat(source, 3, axis=2)
        image = Image.fromarray(source.astype('uint8')).convert('RGB')
    
    else:
        raise TypeError("Input must be a file path (str) or image array (np.ndarray)")

    # Apply validation transform and add batch dimension
    input_tensor = val_transform(image).unsqueeze(0).to(device)

    # Run inference without computing gradients
    with torch.no_grad():
        outputs = model(input_tensor)
        probs = F.softmax(outputs, dim=1)
        confidence, predicted = torch.max(probs, 1)

    # Return label only if confidence is above threshold
    if confidence.item() < threshold:
        return unknown_label
    else:
        return class_names[predicted.item()]
```

## Visualize predict result

This function visualizes a list of images with their predicted and actual labels. Predictions are shown in green if correct and red if incorrect. The output is saved as a PNG file and displayed in a grid.

```python
def visulize(paths, real_label, threshold=0.2):
    plt.figure(figsize=(15, 8))  # Create a large figure for displaying images

    for i, image_path in enumerate(paths):
        # Predict label for each image
        pred = predict_image(image_path, threshold)
        is_correct = (pred == real_label)  # Check if prediction matches real label

        # Read image with OpenCV and convert BGR to RGB for correct color display
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Plot image in a grid with 5 images per row
        plt.subplot(len(paths) // 5 + 1, 5, i + 1)
        plt.imshow(image)
        plt.axis('off')
        # Title shows predicted and real label; green if correct, red if wrong
        plt.title(f"Predict: {pred}\nReal label: {real_label}", color='green' if is_correct else 'red')

    plt.savefig(os.path.join(SAMPLE_FOLDER, f"Predict_{real_label}.png"))  # Save the figure
    plt.tight_layout()
    plt.show()  # Display all images
```
