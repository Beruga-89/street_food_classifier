## 1. Import and setup

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import pandas as pd
import shutil
from tqdm.auto import tqdm
```

```python
device= "cuda" if torch.cuda.is_available() else "cpu"
```

## 2. Data prep

```python
foods_df = pd.read_csv("/kaggle/input/popular-street-foods/popular_street_foods/dataset_stats.csv")
print("Classes:", foods_df['class'].nunique())
class_names = foods_df['class']
```

### 2.1 Train/Val directories

```python
import os
import random
from pathlib import Path

# Paths
source_dir = Path("/kaggle/input/popular-street-foods/popular_street_foods/dataset")
target_base_dir = Path("/kaggle/working/popular_street_foods/data")
train_dir = target_base_dir / "train"
val_dir = target_base_dir / "val"

# val split
val_split = 0.2
random.seed(42)

# create target directory
for split_dir in [train_dir, val_dir]:
    split_dir.mkdir(parents=True, exist_ok=True)

# per class
for class_dir in source_dir.iterdir():
    if class_dir.is_dir():
        images = list(class_dir.glob("*.*"))

        val_count = int(len(images) * val_split)
        val_images = images[:val_count]
        train_images = images[val_count:]

        # create target dir per class
        (train_dir / class_dir.name).mkdir(parents=True, exist_ok=True)
        (val_dir / class_dir.name).mkdir(parents=True, exist_ok=True)

        # copy
        for img_path in train_images:
            shutil.copy(img_path, train_dir / class_dir.name / img_path.name)
        for img_path in val_images:
            shutil.copy(img_path, val_dir / class_dir.name / img_path.name)

        print(f"{class_dir.name}: {len(train_images)} train, {len(val_images)} val")

train_dir = target_base_dir / "train"
val_dir = target_base_dir / "val"
```

### 2.2 Setup transform

```python
# Get pretrained model weights
weights = torchvision.models.EfficientNet_B0_Weights.DEFAULT

# Get transforms used to create pretrained weights
auto_transforms = weights.transforms()
```

```python
train_dataset = datasets.ImageFolder(train_dir, transform=auto_transforms)
val_dataset = datasets.ImageFolder(val_dir, transform=auto_transforms)

train_dataset, val_dataset
```

```python
train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=32, shuffle=False)

train_dataloader, val_dataloader
```

```python
# Preview images
images, labels = next(iter(train_dataloader))
print("Batch image tensor shape:", images.shape)  # [32, 3, 128, 128]
print("Batch label shape:", labels.shape)         # [32]

# Get class names
class_names = train_dataset.classes

# Visualize
def show_batch(img_batch, label_batch):
    plt.figure(figsize=(12, 4))
    for i in range(6):
        img = img_batch[i].permute(1, 2, 0)  # Convert from [C, H, W] to [H, W, C]
        plt.subplot(1, 6, i + 1)
        plt.imshow(img.numpy())
        plt.title(class_names[label_batch[i]])
        plt.axis('off')
    plt.tight_layout()
    plt.show()

show_batch(images, labels)
```

```python
model = torchvision.models.efficientnet_b0(weights=weights).to(device)
```

```python
from torchinfo import summary



# Print a summary using torchinfo (uncomment for actual output)
summary(model=model, 
        input_size=(32, 3, 224, 224), # make sure this is "input_size", not "input_shape"
        # col_names=["input_size"], # uncomment for smaller output
        col_names=["input_size", "output_size", "num_params", "trainable"],
        col_width=20,
        row_settings=["var_names"]
)
```

```python
for param in model.features.parameters():
    param.requires_grad = False
```

```python
torch.manual_seed(42)
torch.cuda.manual_seed(42)

output_shape = len(class_names)

model.classifier = torch.nn.Sequential(
    torch.nn.Dropout(p=0.2, inplace=True),
    torch.nn.Linear(in_features=1280,
                   out_features=output_shape,
                   bias=True)
).to(device)
```

```python
summary(model, 
        input_size=(32, 3, 224, 224), # make sure this is "input_size", not "input_shape" (batch_size, color_channels, height, width)
        verbose=0,
        col_names=["input_size", "output_size", "num_params", "trainable"],
        col_width=20,
        row_settings=["var_names"]
)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 5

for epoch in tqdm(range(epochs)):
    model.train()
    train_loss, train_acc = 0.0, 0.0
    for images, labels in train_dataloader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad
        outputs = model(images)
        loss = criterion(outputs, labels)
        train_loss += loss.item()
        loss.backward()
        optimizer.step()

        outputs_class = torch.argmax(torch.softmax(outputs, dim=1), dim=1)
        train_acc += (outputs_class == labels).sum().item()/len(labels)

    train_loss = train_loss / len(train_dataloader)
    train_acc = train_acc / len(train_dataloader)

    model.eval()
    val_loss, val_acc = 0.0, 0.0
    with torch.inference_mode():
        for images, labels in val_dataloader:
            images, labels = images.to(device), labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item()

            outputs_class = torch.argmax(torch.softmax(outputs, dim=1), dim=1)
            val_acc += (outputs_class == labels).sum().item()/len(labels)
    
    val_loss = val_loss / len(train_dataloader)
    val_acc = val_acc / len(train_dataloader)
            
    print(
      f"Epoch: {epoch+1} | "
      f"train_loss: {train_loss:.4f} | "
      f"train_acc: {train_acc:.4f} | "
      f"val_loss: {val_loss:.4f} | "
      f"val_acc: {val_acc:.4f}"
    )
```
